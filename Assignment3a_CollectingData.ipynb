Now, it is important to isolate certain categories to discover outliers and averages. In order to formulate further research questions when evaluating data related to U.S. crime rates and types, certain questions need to be explored related to these outliers and averages.

The first question to determine outliers is: Which cities have the highest crimes_percapita in 1975 v.s. 2015?
(This will help to determine if there is a difference in the top cities with the highest crime rates per capita in the beginning of the sample dataset and in the last year data was available in the sample dataset.)

```python
crime_df[crime_df['crimes_percapita'] > 1500]
```

Now, I will isolate a city that remains within the cities with the highest crimerates in order to explore it further.

```python
baltimore = crime_df[crime_df['agency_jurisdiction'] == 'Baltimore, MD']
```

What years had the highest crimes_percaptia in Baltimore, MD?

```python
highest_crime_year_baltimore = baltimore.sort_values(by='crimes_percapita', ascending=False)
highest_crime_year_baltimore[:20]
```

Now, we can create a table that displays the years that the robbery rates were the highest in Baltimore, MD, for instance, and ask the question: 

Which years were robbery rates the highest in Baltimore, MD from 1975 - 2015?

```python
baltimore_robberies = baltimore[['report_year', 'robberies']]
top_20_robberies = baltimore_robberies.sort_values(by='robberies', ascending=False)[:20]
top_20_robberies[:20]
```

That looks great, but let's visualize this better with a bar graph to see the fluctuations over time.

```python
import matplotlib.pyplot as plt

top_20_robberies = baltimore_robberies.sort_values(by='robberies', ascending=False)[:20]

plt.bar(top_20_robberies['report_year'], top_20_robberies['robberies'], color='skyblue')
plt.title('Top 20 Years with Highest Robbery Rates in Baltimore')
plt.xlabel('Year')
plt.ylabel('Robberies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```
